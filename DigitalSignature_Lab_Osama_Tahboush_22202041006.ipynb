```python
# DigitalSignature_Lab_<YourName>_<EnrollNo>.ipynb
# Special Topics in Information Security – Lab 03
# Implementation of ECDSA Digital Signature System

# Step 1: Import libraries
from ecdsa import SigningKey, SECP256k1
import hashlib

# Step 2: Simulated transaction message
message = "Ahmed pays 100 coins to Ali"
print("Transaction Message:", message)

# Step 3: Generate ECDSA private-public key pair
private_key = SigningKey.generate(curve=SECP256k1)
public_key = private_key.get_verifying_key()

print("\nPrivate Key:", private_key.to_string().hex())
print("Public Key:", public_key.to_string().hex())

# Step 4: Hash the message using SHA-256
message_hash = hashlib.sha256(message.encode()).digest()
print("\nSHA-256 Hash:", hashlib.sha256(message.encode()).hexdigest())

# Step 5: Sign the hash with private key
signature = private_key.sign(message_hash)
print("\nSignature:", signature.hex())

# Step 6: Verify the signature with public key
is_verified = public_key.verify(signature, message_hash)
print("Signature Verified:", is_verified)
```
